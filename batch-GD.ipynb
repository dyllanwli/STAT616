```python
all_loss = []
all_step = []
last_a = a
last_b = b

for step in range(1,100):
    loss = 0
    all_da = 0
    all_db = 0
    for i in range(0, len(x)):
        y_p = a*x[i] + b
        loss = loss + (y[i] - y_p)*(y[i] - y_p)/2
        all_da = all_da + utils.da(y[i], y_p, x[i])
        all_db = all_db + utils.db(y[i], y_p)
    loss = loss/len(x)

    # draw fig.1 contour line
    plt.subplot(1, 2, 1)
    plt.scatter(a, b, s = 5, color='black')
    plt.plot([last_a, a], [last_b, b], color='blue')

    # draw fig.2 loss line
    all_loss.append(loss)
    all_step.append(step)
    plt.subplot(1, 2, 2)
    plt.plot(all_step, all_loss, color='orange')
    plt.title('BGD')
    plt.xlabel("step")
    plt.ylabel("loss")

    # update param
    last_a = a
    last_b = b
    a = a - learning_rate * all_da
    b = b - learning_rate * all_db

    if step % 5 == 0:
        print("step: ", step, " loss: ", loss, " a: ", a, " b: ", b)
        plt.show()
        plt.pause(0.01)

plt.show()
```
