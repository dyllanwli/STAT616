```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import pandas as pd 

import seaborn as sns

from sklearn import preprocessing
```

```python
diabetes = load_diabetes()
data = pd.DataFrame(diabetes.data)
data.columns = diabetes.feature_names
data['disease'] = diabetes.target 

col = list(diabetes.feature_names)
col.append('disease')
```

```python
normalized = preprocessing.normalize(data.values, norm="l1", axis=0)
```

```python
data = pd.DataFrame(normalized, columns=col)

# Finding out the correlation between the features
corr = data.corr()
```

```python
# Plotting the heatmap of correlation between features
plt.figure(figsize=(20,20))
sns.heatmap(corr, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='Greens')
```

```python
data.to_csv("normalized_diabetes.csv", index=False)
```
